### Data Cleaning

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.rename(str.strip,axis='columns',inplace=True)
df.columns
```

```python
for col in df.columns:
    if df[col].isnull().sum()>0:
        val = df[col].mean()
        df[col] = df[col].fillna(val)
df.isnull().sum()
```

```python
df.replace({'yes':1,'no':0},inplace=True)
```

```python
plt.figure(figsize=(10,10))
sb.heatmap(df.corr()>0.8,annot=True,cbar=False)
plt.show()
```

```python
df.drop(['maxtemp', 'mintemp'], axis=1, inplace=True)
```

```python
features = df.drop(['day', 'rainfall'], axis=1)
target = df.rainfall
```

```python
X_train, X_test, \
    Y_train, Y_test = train_test_split(features,
                                      target,
                                      test_size=0.2,
                                      stratify=target,
                                      random_state=2)
ros = RandomOverSampler(sampling_strategy='minority',random_state=22)
X , Y  = ros.fit_resample(X_train,Y_train)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_test = scaler.transform(X_test)
```

```python
models = [LogisticRegression(),XGBClassifier(),SVC(kernel='rbf',probability=True)]

for i in range(3):
    models[i].fit(X,Y)
    print(f'{models[i]} : ')
    train_preds = models[i].predict_proba(X)
    print('Training Accuracy : ',metrics.roc_auc_score(Y,train_preds[:,1]))
    val_preds = models[i].predict_proba(X_test)
    print('Validation Accuracy : ', metrics.roc_auc_score(Y_test,val_preds[:,1]))
    print()
```

```python
ConfusionMatrixDisplay.from_estimator(models[2],X_test,Y_test)
plt.show()
```

```python
print(metrics.classification_report(Y_test,models[2].predict(X_test)))
```
